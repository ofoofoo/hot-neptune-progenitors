```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
bee = []
for i in range(20):
    print("multiplicity_" + str(i)+".pickle")
    try:
        with (open("/home/orioncfoo/Hot Neptune Project/0622 TOI 969 Run 1/orbitalelementshash_" + str(i)+".pickle", "rb")) as openfile:
            bee.append(pickle.load(openfile))
            
    except:
        print("run not finished yet!")
```

```python
print(len(bee))
print(bee[0])
print(np.linspace(0, 10000000, 1000))
```

```python
for elm in bee:
    fig = plt.figure(figsize=(20,10))
    ax = plt.subplot(221)
    ax.set_xlabel("time")
    ax.set_ylabel("semi-major axis")
    ax.set_title("Semi-Major Axis")
    ax.set_yscale("log")
    #plt.xlim(0, 2000)
    #print(times)
    count = 1
    for i in elm.keys():
        plt.plot(elm[i][0][0], elm[i][0][1], label = "Planet " + str(count))
        count += 1
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
print("bob)")
```
